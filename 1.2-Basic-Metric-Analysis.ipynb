# Basic metric analysis

The following graphs represents the basic metrics analysis for domain1_score calculated for each essay set (1-8)
    - mean 
    - median
    - mode
    - standard deviation 
    - variance
    

```python
import nltk
import pandas as pd
#df = pd.read_csv('Automated-Essay-Scoring-master/essays_and_scores.csv' , encoding = 'latin-1')
data = pd.read_csv('kaggle-dataset/training_set_rel3.tsv', sep='\t', encoding = "ISO-8859-1", header=0)

# getting relevant columns
df = data[['essay_id','essay_set','rater1_domain1','rater2_domain1','domain1_score']].copy()

df
```

```python
df.mean(axis=0)
```

```python
df.median(axis=0) #Median of each essay set
```

```python
df.mode(axis=0) #Mode of each essay set
```

```python
df.std() #standard deviation of whole csv file
```

```python
df.var()    #variance of whole csv file
```

```python
df.groupby('essay_set').domain1_score.mean()  #Mean of each essay set

```

```python
df.groupby('essay_set').domain1_score.median() #Median of each essay set
```

```python
df.groupby('essay_set').domain1_score.std() #standard deviation of each essay set
```

```python
df.groupby('essay_set').domain1_score.var() #variance per essay set
```

```python
import matplotlib.pyplot as plt
df.groupby('essay_set').domain1_score.mean().plot.bar(y='essay', rot=0, legend=False)
plt.title('Mean with respect to each essay set')
plt.ylabel('Mean')
plt.show()
```

```python
import matplotlib.pyplot as plt
df.groupby('essay_set').domain1_score.median().plot.bar(y='essay', rot=0, legend=False)
plt.title('Median with respect to each essay set')
plt.ylabel('Median')
plt.show()
```

```python
import matplotlib.pyplot as plt
df.groupby('essay_set').domain1_score.std().plot.bar(y='essay', rot=0, legend=False)
plt.title('Std deviation with respect to each essay set')
plt.ylabel('Standard Deviation')
plt.show()
```

```python
import matplotlib.pyplot as plt
df.groupby('essay_set').domain1_score.var().plot.bar(y='essay', rot=0, legend=False)
plt.title('Variance with respect to each essay set')
plt.ylabel('Variance')
plt.show()
```
